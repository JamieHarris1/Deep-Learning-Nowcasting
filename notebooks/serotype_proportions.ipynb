```python
import pandas as pd
from pathlib import Path
import sys
import os
import numpy as np


sys.path.append(os.path.abspath('../src'))
project_dir = project_dir = Path.cwd().parent

# GISAID settings:
# Location: South America, Brazil
# Collection: 2012-12-31 -> 2020-12-31
# Total as off 23/07/25 is 1509

serotype_props = pd.read_csv(project_dir / "data" / "raw" / "Serotype_Progression.csv")
serotype_props
```

```python
serotype_prop_df = pd.DataFrame({
        "Date": [],
        "Prop": [],
        "Serotype": []

    })
for i,serotype in enumerate(serotype_props['variant']):
    dates = serotype_props['yearmth'].iloc[i].split(',')
    prop = np.array(serotype_props['percentage'].iloc[i].split(','), dtype=np.float32)

    new_df = pd.DataFrame({
        "Date": dates,
        "Prop": prop,
        "Serotype": serotype

    })
    serotype_prop_df = pd.concat([serotype_prop_df, new_df])

serotype_prop_df = serotype_prop_df.pivot(index='Date', columns='Serotype', values='Prop')
serotype_prop_df
```

```python
serotype_prop_df.to_csv(project_dir / "data" / "transformed" / 'serotype_prop.csv', index=True)
```
