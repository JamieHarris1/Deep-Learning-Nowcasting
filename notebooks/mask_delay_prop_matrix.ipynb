```python
import sys
import os
import pandas as pd
import numpy as np
sys.path.append(os.path.abspath('../src'))
```

```python
max_delay = 40
t = max_delay + 1
delay_prop_df = pd.read_csv("../data/transformed/dengue-sp-reporting-delay-proportion.csv", index_col=0)
delay_prop_matrix = np.array(delay_prop_df.iloc[t-max_delay:t])

for i in range(max_delay):
    for j in range(max_delay):
        if i + j > max_delay - 1:
            delay_prop_matrix[i, j] = 0

pd.DataFrame(delay_prop_matrix)
```
