```python
class ReportingDataset():
    def __init__(self, df, D):
        self.df = df.copy()
        self.D = D
        self.max_val = self.df.values.max()
    
    def get_obs(self, idx, n_points, norm=False, mask=False):
        # Ignore first D days to ensure full matrix
        t = idx + self.D


        # Create reporting triangle for day t
        matrix = self.df.iloc[t - n_points + 1: t + 1, :self.D].values
        
        # Mask last D days
        if mask:
            matrix[-self.D:] = self.mask_matrix(matrix[-self.D:])

        if norm:
            return matrix / self.max_val
        else:
            return matrix
    
    def get_y(self, idx):
        t = idx + self.D
        return self.df.iloc[t].sum()

    def get_date(self, idx):
        t = idx + self.D
        return self.df.index[t]


    def get_past_triangles(self, idx, n_points=1, norm=True):
        assert n_points >= 1, "n must be non-negative"
        assert idx + 1 >= n_points, "idx must be at least as big as number of points"
        triangles = np.zeros(shape=(n_points, self.D, self.D))
        t_vals = np.arange(idx-n_points+1, idx+1)
        for i, t in enumerate(t_vals):
            triangles[i, :, :] = self.get_obs(t, n_points=self.D, norm=norm)

        return triangles

    def mask_matrix(self, matrix):
        for i in range(self.D):
            for j in range(self.D):
                if i + j > self.D - 1:
                    matrix[i, j] = 0
        return matrix

train_obj = ReportingDataset(train_df, D=trunc_D)
```

```python
def silu(x):
    return x * pm.math.sigmoid(x)


def sampler_kwargs():
    return dict(
        nuts_sampler="nutpie",
        chains=2,
        draws=300,
        tune=1000,
        target_accept=0.95,
        max_treedepth=15,
        nuts_sampler_kwargs={"backend": "jax", "gradient_backend": "jax"}
    )

def get_train_post(var, train_post):
    return np.percentile(train_post[var], 50, axis=-1)

def BNN_nowcast(data_obj, t, n_points, progressbar=True, train_post=False):

    if train_post:
        obs = data_obj.get_obs(t, n_points, mask=True)
    else:
        obs = data_obj.get_obs(t, n_points)

    triangles = data_obj.get_past_triangles(t, n_points)
    



    t = np.arange(0, n_points)
    t_week = t % 7                      # day of the week number

    # Global trend spline with 6 basis funcs and 3 degree polynomial
    spline_trend = dmatrix(
        "bs(t, df=10, degree=3, include_intercept=False)", {"t": t}, return_type='dataframe'
    )
    X_trend = np.asarray(spline_trend)

    # cc is syclic cubic spline
    spline_week = dmatrix(
        "cc(t_week, df=6)", {"t_week": t_week}, return_type='dataframe'
    )
    X_week = np.asarray(spline_week)


    with pm.Model() as model:

        # Trend spline coefficients
        sigma_trend = pm.HalfNormal("sigma_trend", 3)
        beta_trend = pm.Normal("beta_trend", mu=0, sigma=sigma_trend, shape=X_trend.shape[1])
        
        # Cyclic spline coefficients (seasonality)
        sigma_week = pm.HalfNormal("sigma_week", 3)
        beta_week = pm.Normal("beta_week", mu=0, sigma=sigma_week, shape=X_week.shape[1])
        
        log_lam = pm.Deterministic(
            "log_lam", pm.math.dot(X_trend, beta_trend) + pm.math.dot(X_week, beta_week)
        )
        lam = pm.Deterministic("lam", pm.math.exp(pm.math.clip(log_lam, -5, 10)))


        # Neural net proportions
        T, D, F = triangles.shape
        n_hidden = 8
        

        h0 = pm.math.flatten(triangles, ndim=2)  # (T, D*F)
        
        if train_post:
            net_sd = 0.1
            W1 = pm.Normal("W1", get_train_post("W1", train_post), net_sd, shape=(D*F, n_hidden))
            b1_net = pm.Normal("b1_net", get_train_post("b1_net", train_post), net_sd, shape=(n_hidden,))
            h1 = silu(pm.math.dot(h0, W1) + b1_net)

            W2 = pm.Normal("W2", get_train_post("W2", train_post), net_sd, shape=(n_hidden, D))
            b2_net = pm.Normal("b2_net", get_train_post("b2_net", train_post), net_sd, shape=(D,))
            p_raw = pm.math.dot(h1, W2) + b2_net

        else: 
            net_sd = 0.3
            W1 = pm.Normal("W1", 0, net_sd, shape=(D*F, n_hidden))
            b1_net = pm.Normal("b1_net", 0, net_sd, shape=(n_hidden,))
            h1 = silu(pm.math.dot(h0, W1) + b1_net)

            W2 = pm.Normal("W2", 0, net_sd, shape=(n_hidden, D))
            b2_net = pm.Normal("b2_net", 0, net_sd, shape=(D,))
            p_raw = pm.math.dot(h1, W2) + b2_net


        p_raw = pm.math.clip(p_raw, -10, 10)
        p = pm.Deterministic("p", pm.math.softmax(p_raw, axis=1))

        mu = pm.Deterministic("mu", lam[:, None] * p)
        mu_clipped = pm.math.clip(mu, 1e-3, 1e4)

        alpha = pm.Exponential("alpha", 2)
        z = pm.NegativeBinomial("z", mu_clipped, alpha, observed=obs)

        idata = pm.sample(progressbar=progressbar, **sampler_kwargs())
        return idata

def get_PNN_nowcast(triangles, D):
    

    nowcast_pnn = NowcastPNN(past_units=D, max_delay=D, conv_channels=[16, 1], hidden_units=[16, 8], dropout_probs=[0.3, 0.1])
    nowcast_pnn.load_state_dict(torch.load(f"../src/outputs/weights/weights-{D}-{D}"))

    # Put model in eval mode with dropout active
    nowcast_pnn.eval()
    nowcast_pnn.drop1.train()
    nowcast_pnn.drop2.train()

    n_samples = 1000
    preds = np.zeros(shape=(len(triangles), n_samples))
    for i in range(n_samples):
        matrix = torch.tensor(triangles)
        preds[:, i] = nowcast_pnn(matrix).sample().numpy()
        

    PNN_preds = np.quantile(preds, 0.5, axis=1)
    PNN_lower = np.percentile(preds, 2.5, axis=1)
    PNN_upper = np.percentile(preds, 97.5, axis=1)
    return PNN_preds, PNN_lower, PNN_upper
```
